```python
import pandas as pd
df = pd.read_excel('./분류된_상품_데이터.xlsx')
df
```

```python
import matplotlib.pyplot as plt
```

```python
d = df[df['상품 분류'] == '일반 상품']
d
```

```python
f = df[df['상품 분류'] == '교육용 상품']
f
```

```python
# 주문일자를 날짜 형식으로 변환
df['주문일자'] = pd.to_datetime(df['주문일자'])

# 월별로 주문 데이터를 그룹화
df['월'] = df['주문일자'].dt.to_period('M')

# 각 고객이 각 상품을 몇 번 구매했는지 계산
구매빈도1 = d.groupby(['업체명', '상품명', '월']).size().unstack().fillna(0)

# 월별로 재구매한 상품과 처음 구매한 상품 개수를 구분
재구매1 = (구매빈도1 > 1).sum()
처음구매1 = (구매빈도1 == 1).sum()

# 월별 재구매율 계산
재구매율_월별1 = (재구매1 / (처음구매1 + 재구매1)) * 100


print("월별 일반상품 재구매율:")
print(재구매율_월별)

# 월 데이터를 문자열로 변환
재구매율_월별1.index = 재구매율_월별1.index.strftime('%Y-%m')


# 그래프 그리기
#재구매율_월별.plot()
# 그래프 그리기
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 6))
plt.plot(재구매율_월별1.index, 재구매율_월별1.values, marker='o', linestyle='-')
plt.xlabel('월')
plt.ylabel('재구매율 (%)')
plt.title('월별 일반상품 재구매율')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# 주문일자를 날짜 형식으로 변환
df['주문일자'] = pd.to_datetime(df['주문일자'])

# 월별로 주문 데이터를 그룹화
df['월'] = df['주문일자'].dt.to_period('M')

# 각 고객이 각 상품을 몇 번 구매했는지 계산
구매빈도2 = f.groupby(['업체명', '상품명', '월']).size().unstack().fillna(0)

# 월별로 재구매한 상품과 처음 구매한 상품 개수를 구분
재구매2 = (구매빈도2 > 1).sum()
처음구매2 = (구매빈도2 == 1).sum()

# 월별 재구매율 계산
재구매율_월별2 = (재구매2 / (처음구매2 + 재구매2)) * 100


print("월별 교육용상품 재구매율:")
print(재구매율_월별2)

# 월 데이터를 문자열로 변환
재구매율_월별2.index = 재구매율_월별2.index.strftime('%Y-%m')


# 그래프 그리기
#재구매율_월별.plot()
# 그래프 그리기
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 6))
plt.plot(재구매율_월별2.index, 재구매율_월별2.values, marker='o', linestyle='-',label ='교육용상품')
plt.xlabel('월')
plt.ylabel('재구매율 (%)')
plt.title('월별 상품 재구매율')
plt.grid(True)
plt.xticks(rotation=45)
plt.plot(재구매율_월별1.index, 재구매율_월별1.values, marker='o', linestyle='-', label = '일반상품')
plt.legend()
plt.show()
```

```python
### 일반상품 재구매율 ###
# 월별로 고객의 재구매율 계산
d['월'] = d['주문일자'].dt.to_period('M')
재구매율_일 = d.groupby(['업체명', '상품명', '월'])['주문일자'].transform('count') > 1

# 재구매율을 월별로 그룹화하여 계산
월별_재구매율_일 = d[재구매율_일].groupby('월')['월'].count() / d.groupby('월')['월'].count() * 100
```

```python
# 주문일자를 날짜 형식으로 변환
df['주문일자'] = pd.to_datetime(df['주문일자'])

# 월 데이터를 문자열로 변환
월별_재구매율_일.index = 월별_재구매율_일.index.strftime('%Y-%m')

# 그래프 그리기
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(10, 6))
plt.plot(월별_재구매율_일.index, 월별_재구매율_일.values, marker='o', linestyle='-')
plt.xlabel('월')
plt.ylabel('재구매율 (%)')
plt.title('월별 일반상품 재구매율')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# '상품 분류' 별로 '판매금액'과 '주문수량'을 그룹화하고 합계를 계산
ag = df.groupby('상품 분류')[['판매금액', '주문수량']].sum()
print(ag)

```
